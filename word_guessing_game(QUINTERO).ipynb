```python
import random

def select_random_word():
    with open('words.txt') as file:
        words_list = file.read().splitlines()
    return random.choice(words_list)

def display_word(word, guessed_letters):
    return ' '.join([letter if letter in guessed_letters else '_' for letter in word])

def get_guess():
    while True:
        guess = input('Enter a letter: ')
        if len(guess) == 1 and guess.isalpha():
            return guess.lower()
        print('Invalid input. Please enter a single letter.')

def play_word_game():
    remaining_guesses = 7
    chosen_word = select_random_word()
    guessed_letters = set()
    
    print('GUESS THE WORD TO WIN THE GAME!')
    print('The word has', len(chosen_word), 'letters.')
    
    while remaining_guesses > 0:
        print(display_word(chosen_word, guessed_letters))
        print('You have', remaining_guesses, 'chances.')
        
        guessed_letter = get_guess()
        if guessed_letter in guessed_letters:
            print('You already chose that letter, can you chose another letter?.')
        elif guessed_letter in chosen_word:
            guessed_letters.add(guessed_letter)
            print('NICE GUESS!KEEP IT UP')
            if set(chosen_word) == guessed_letters:
                print('Congratulations, YOU WON!')
                return
        else:
            guessed_letters.add(guessed_letter)
            remaining_guesses -= 1
            print('Sorry the letter you select are not in the word that you guessing.')
    
        if remaining_guesses == 0:
            print('Game Over!! Sorry, but your chances are already run out', chosen_word)
            break
    
play_word_game()
```
